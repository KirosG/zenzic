```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# display plots in the notebook
%matplotlib inline
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (14, 8)
plt.rcParams['font.size'] = 14
```

```python
# data source https://www.theatlas.com/charts/E1Wxox0c
raw_data = {
    'Year':[2005,2007,2009,2010,2011,2013,2015,2017,2019,2021],
    'Data':[130,281,800,1227,1800,4400,7910,15820,31640,63280]
}
```

```python
df = pd.DataFrame(raw_data,columns=['Data'],index=raw_data['Year'])
```

```python
df.index = pd.to_datetime(df.index, format='%Y')
```

```python
df.head(10)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# plot data
ax.plot(df.Data, color='#E9212E', linewidth=4)

# labels
ax.set_xlabel('Year',  color='black', size=20)
ax.set_ylabel('Exabytes',  size=20)

# titles
plt.suptitle("Big Data is Getting Bigger", y=0.96, size=20)
plt.title("Total World Data Over Time", y=1.01, size=14)

# citation
ax.text(1, -0.15, 'Data: IDC Digital Universe (2007-2014)\n Chart: Represently',
        horizontalalignment='right',
        verticalalignment='bottom',
        transform=ax.transAxes,
        fontsize=8)
plt.savefig('./graphs/big-data.png', dpi=300, bbox_inches='tight')
```

```python
#iPhoneX reference point

fig = plt.figure()
ax = fig.add_subplot(111)

# plot data
ax.plot(df.Data, color='#E9212E', linewidth=4)

# labels
ax.set_xlabel('Year',  color='black', size=20)
ax.set_ylabel('Exabytes',  size=20)

# titles
plt.suptitle("Big Data is Getting Bigger", y=0.96, size=20)
plt.title("Total World Data Over Time", y=1.01, size=14)

# iPhone reference
plt.vlines(x='2013', ymin=0, ymax=max(df.Data)*.85, linewidth=2, linestyle=":", color='#2D68AF')
ax.text(0.48, 0.80, "2013: So Much Data...",
        horizontalalignment='right',
        verticalalignment='bottom',
        transform=ax.transAxes,
        fontsize=14,
        fontweight='bold')

ax.text(0.48, 0.72, "If stored on the iPhoneX, everyone in\nthe world could hold one in each hand",
        horizontalalignment='right',
        verticalalignment='bottom',
        transform=ax.transAxes,
        fontsize=12)

# citation
ax.text(1, -0.15, 'Data: IDC Digital Universe (2007-2014)\n Chart: Represently',
        horizontalalignment='right',
        verticalalignment='bottom',
        transform=ax.transAxes,
        fontsize=8)
plt.savefig('./graphs/big-data-iphone-caption.png', dpi=300, bbox_inches='tight')
```
